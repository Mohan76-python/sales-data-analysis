# üìä Sales Data Analysis

## üß† Objective
Analyze sales data to uncover trends and performance insights using Pandas and Matplotlib.

## üìÇ Dataset Overview
- Columns: Date, Product, Region, Sales
- No missing values
- 4 products across 4 regions

## üìà Key Insights
- Widget A had the highest total sales.
- North region contributed the most revenue.
- Sales peaked in early January.

## üõ† Tools Used
- Python 3
- Pandas
- Matplotlib
- Jupyter Notebook



## üìà Insight: Sales by Product
Widget A is the top-selling product, followed by Widget B. Widget C has the lowest sales.
## üîç Insight: Sales by Region
- North region contributes 45% of total sales.
- South region has the lowest sales ‚Äî potential for growth or investigation.
## üí° Recommendations
- Increase marketing in South region to boost sales.
- Promote Widget C with discounts or bundles.
- Investigate why sales dipped in March.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('sales_data.csv')  


sns.set(style="whitegrid")  
product_sales = df.groupby('Product')['Sales'].sum()

product_sales.plot(kind='bar', figsize=(8,5), color='skyblue')
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
region_sales = df.groupby('Region')['Sales'].sum()

region_sales.plot(kind='pie', autopct='%1.1f%%', figsize=(6,6), title='Sales Distribution by Region')
plt.ylabel('')  
plt.show()
df['Date'] = pd.to_datetime(df['Date'])
daily_sales = df.groupby('Date')['Sales'].sum()

daily_sales.plot(kind='line', marker='o', figsize=(10,5))
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
df['Month'] = df['Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales.plot(kind='bar', figsize=(8,5), color='orange')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
top_product = product_sales.idxmax()
top_sales = product_sales.max()

print(f"Top-selling product: {top_product} with ‚Çπ{top_sales}")
daily_sales.rolling(window=7).mean().plot()

```
